```python
#create a program to compute student marks average and classification for example >90 A grade 80-89 B grade 70-79 C grade 60-69 D grade below 60 F grade
maths=int(input("maths marks:"))
science=int(input("science marks:"))
english=int(input("english marks:"))
hindi=int(input("hindi marks:"))
avg=(maths+science+english+hindi)/4
print(avg)
if avg>90:
    print("A grade")
elif avg>=80 and avg<90:
    print("B grade")
elif avg>=70 and avg<80:
    print("C grade")
elif avg>=60 and avg<70:
    print("D grade")
else:
    print("F grade")
```

```python
#calculator
num_1=int(input())
num_2=int(input())
op=input("enter operation:")
if op=="+":
    print(num_1+num_2)
elif op=="-":
    print(num_1-num_2)
elif op=='*':
  print(num_1*num_2)
elif op=="/":
  print(num_1/num_2)
else:
  print("error")
```

```python
#factorial of a number
num=int(input())
fact=1
for i in range(1,num+1):
  fact=fact*i
print(fact)
```

```python
#fibonacci
num=int(input())
a=0
b=1
print(a)
print(b)
for i in range(2,num):
  c=a+b
  a=b
  b=c
  print(c)
```

```python
#in an ac circuit the total impedence z of a series rlc circuit z=r+j(wl-1/wc) w=2pif 1. your task is to compute the impedence as a compelx number 2. return its magnitude and phase angle is degree input(r=100 l=0.5 c=100e^-6 f=15)
import math
import cmath
r=100
l=0.5
c=100*(math.e**(-6))
f=15
w=2*math.pi*f
img=(w*l-(1/(w*c)))
a=complex(r,img)
print(a)
mag=math.sqrt(a.real**2+a.imag**2)
print(mag)
phase=math.degrees(cmath.phase(a))
print(phase)
```
